```python
import pandas as pd
from ast import literal_eval
from utils import specializations


task = 'Primary specializations review'
column = 'primary_specializations'
column_by_review = 'reviewed_' + column

evaluate_file = './data/' + task + '.csv'
evaluate_df = pd.read_csv(
    evaluate_file,
    converters={
        column: literal_eval,
        column_by_review: literal_eval,
    },
)

normalize = lambda row: set([spec['id'] for spec in row if spec['score'] >= 40])
spec = evaluate_df[column].apply(normalize)
spec_by_review = evaluate_df[column_by_review].apply(normalize)

find_common = lambda set1, set2: set1.intersection(set2)
common_spec = spec.combine(spec_by_review, find_common)

spec_to_count = [spec, spec_by_review, common_spec]
count = []
for item in spec_to_count:
    item = item.astype(str)
    total = []
    for spec in specializations:
        num = item.str.count(spec).sum()
        total.append(num)
    count.append(total)

count = list(map(list, zip(*count)))
print(count)
```

```python
results = pd.DataFrame(
    count,
    columns=['retrieved_instances', 'relevant_instances', 'relevant_retrieved_instances'],
    index=specializations,
)

results['precision'] = results['relevant_retrieved_instances'] / results['retrieved_instances']
results['recall'] = results['relevant_retrieved_instances'] / results['relevant_instances']

results
```

```python
results.to_csv('results.csv')
```
